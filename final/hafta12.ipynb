```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
data.Gender.unique()
```

```python
genders = []
for i,row in data.iterrows():
    if row.Gender == "Female":
        genders.append(0)
    else:
        genders.append(1)
```

```python
ages = data.Age.values
```

```python
data["Marital Status"].unique()
```

```python
maritals = []
for i,row in data.iterrows():
    if row["Marital Status"] =="Single":
        maritals.append(0)
    elif row["Marital Status"] =="Married":
        maritals.append(1)
    else:
        maritals.append(2)
```

```python
len(maritals)
```

```python
data.Occupation.unique()
```

```python
occupations = []
for i,row in data.iterrows():
    if row.Occupation == "Student":
        occupations.append(0)
    elif row.Occupation == "Employee":
        occupations.append(1)
    elif row.Occupation == "Self Employeed":
        occupations.append(2)
    else:
        occupations.append(3)
```

```python
import numpy as np
```

```python
genders = np.array(genders)
ages = np.array(ages)
maritals = np.array(maritals)
egitim = np.stack((ages,genders,maritals),axis=1)
test = np.array(occupations)
```

```python
egitim
```

```python
egitim.shape
```

```python
test.shape
```

```python
X_egitim, X_test, y_egitim, y_test = train_test_split(egitim,test,test_size=0.2,random_state=42) 
```

```python
X_egitim.shape
```

```python
siniflandirici = DecisionTreeClassifier()
siniflandirici.fit(X_egitim, y_egitim)
```

```python
tahminler = siniflandirici.predict(X_test)
```

```python
tahminler
```

```python
y_test
```

```python
dogruluk = accuracy_score(y_test, tahminler)
dogruluk
```
