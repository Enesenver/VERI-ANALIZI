```python
genders = []
for i,row in data.iterrows():
    if row.Gender == "Female":
        genders.append(0)
    else:
        genders.append(1)
```

```python
ages = data.Age.values
```

```python
Departures = data["Departure Delay"].values
```

```python
Arrivals = data["Arrival Delay"].values
```

```python
data["Customer Type"].unique()
```

```python
customers = []
for i,row in data.iterrows():
    if row["Customer Type"] =="Returning": 
        customers.append(0)
    elif row["Customer Type"] =="First-time":
        customers.append(1)
    else:
        customers.append(2)
```

```python
len(customers)
```

```python
Types = []
for i,row in data.iterrows():
    if row["Type of Travel"] =="Business":
        Types.append(0)
    elif row["Type of Travel"] =="Personal":
        Types.append(1)
    else:
        Types.append(2)
```

```python
len(Types)
```

```python
import numpy as np
```

```python
genders = np.array(genders)
ages = np.array(ages)
Departures = np.array(Departures)
Arrivals = np.array(Arrivals)
customers = np.array(customers)
egitim = np.stack((ages,genders,customers,),axis=1)
test = np.array(Types)
```
